# Polynomial Regression

## Importing all the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

## Loading the dataset

```python
df = pd.read_csv('/Users/omkarmutreja/Downloads/Polynomial_Regression/Position_Salaries.csv')
print(df.head())
X = df.iloc[:,1:2].values
y = df.iloc[:,2].values
print(X.shape)
print(y.shape)
```

## Fitting the Linear Regression model to the dataset

```python
# Here, we do not split the data into train and test sets as the dataset is very small
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

## Fitting Polynomial Regression model to the dataset

```python
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly,y)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
```

## Visualizing the Linear Regression results

```python
plt.scatter(X,y,c='red')
plt.plot(X,lin_reg.predict(X),c='blue')
plt.title('Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Visualizing the Polynomial Regression results

```python
plt.scatter(X,y,c='red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),c='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with Linear Regression and Polynomial Regression

```python
lin_pred_value = lin_reg.predict(6.5)
poly_pred_value = lin_reg2.predict(poly_reg.fit_transform(6.5))
print("Linear_reg predicted value: ",lin_pred_value)
print("Polynomial_reg predicted value: ",poly_pred_value)
```

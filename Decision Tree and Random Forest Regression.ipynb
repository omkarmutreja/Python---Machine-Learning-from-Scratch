# Decision Tree Regression and Random Forest

## Import all the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

## Load the dataset

```python
df = pd.read_csv("/Users/omkarmutreja/Downloads/Decision_Tree_Regression/Position_Salaries.csv")
df.head()
X = df.iloc[:,1:2].values
y = df.iloc[:,2].values
print(X.shape)
print(y.shape)
```

## Fitting the Decision Tree Regression to the dataset

```python
reg = DecisionTreeRegressor(random_state=0)
reg.fit(X,y)
reg.predict(6.5)
```

## Visualizing the result (Higher Resoultion and Smoother curve)

```python
X_grid = np.arange(min(X),max(X),0.01)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y,c='red')
plt.plot(X_grid,reg.predict(X_grid),c='blue')
plt.title('Decision Tree Regression for higher resoultion and smoother curve')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Fitting Random Forest Regression to the dataset

```python
reg = RandomForestRegressor(n_estimators=100, random_state=0)
reg.fit(X,y)
reg.predict(6.5)
```

```python
X_grid = np.arange(min(X),max(X),0.01)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y,c='red')
plt.plot(X_grid,reg.predict(X_grid),c='blue')
plt.title('Random Forest Regression for higher resoultion and smoother curve')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
